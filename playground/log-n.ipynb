```python
import matplotlib.pyplot as plt
import numpy as np

# 创建x轴数据
x = np.linspace(1, 1000, 1000)

# 计算y轴数据(log(N))
y = np.log(x)

# 创建图形
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制曲线
ax.plot(x, y)

# 设置标题和坐标轴标签
ax.set_title('log(N) Curve')
ax.set_xlabel('N')
ax.set_ylabel('log(N)')

# 显示图形
plt.show()
```
